```python
# Import the libaries matplotlib and pandas as plt and pd variables respectively
# Matplotlib is a data visualization tool to display plots
# Pandas is a data manipulation tool to manipulate data with built in helper functions

import matplotlib.pyplot as plt
import pandas as pd

"""
The line below is used to render plots from the matplotlib INLINE 
i.e within the cell itself as apposed to a separate screen)
"""
# %matplotlib inline

"""
pd.read_csv takes in a file with extensions such as .data, .csv etc 
and converts to a dataframe using a delimiter/separation ","(comma)
"""
iris =  pd.read_csv("bezdekIris.data", sep=",")

# Add the following columns to the dataframe
iris.columns = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class_name"]

# Create a dictionary of the classname and the color assigned to it
colors = {'Iris-setosa':'r', 'Iris-virginica':'g', 'Iris-versicolor':'b'}

fig, ax = plt.subplots()

for iris_name, iris_group in iris.groupby('class_name'):
    ax.scatter(
        iris_group['sepal_length'], # X axis
        iris_group['sepal_width'],  # Y axis
        color=colors[iris_name], 
        label=iris_name
    )
    
ax.set_title('Iris Dataset')
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

plt.legend()
plt.show()
```

```python
sepal_length_df_average = iris.groupby('class_name')['sepal_length'].mean().to_frame().reset_index()

plt.bar(
    sepal_length_df_average['class_name'],
    sepal_length_df_average['sepal_length'],
    color = ['r','b','g']
) 

plt.title('Mean Sepal Length of each species')
plt.xlabel('Species')
plt.ylabel('Sepal Length') 

plt.show()
```
